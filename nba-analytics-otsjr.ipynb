Importar o dataset de Temporada Regular

Importar o dataset de playoff

```python
#Import playoff dataset
df_PO = pd.read_csv('../input/20222023-nba-player-stats-regular/2022-2023 NBA Player Stats - Playoffs.csv', delimiter = ';', index_col = 'Rk', encoding = 'ISO-8859-1')
```

```python
#Players who scored more than 30 points in the playoffs
df_PO.query('PTS > 30')
```

Gráfico de pontos vs bloqueios na Temporada Regular

```python
#Points vs Blocks in the regular season
plt.scatter(x=df_reg['BLK'], y=df_reg['PTS'], c="g", s=2)
plt.grid()
plt.xlim(0, 50)
plt.ylim(0, 50)
plt.title("Blocks vs Points")
plt.xlabel("Blocks")
plt.ylabel("Point")
plt.legend(loc=2)
plt.show()
```
